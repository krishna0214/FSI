```python


import numpy as np
import time
import threading
import matplotlib.pyplot as plt
from queue import Queue


queue1 = Queue()
queue2 = Queue()
def Graph(u_star, p_star):
    queue1.put(u_star)
    queue2.put(p_star)
    batch_size = 10
    if queue1.qsize()>batch_size:
        
       def plot_data():
            fig = plt.figure()
            start_time = time.time()  # Start the timer
            while True:
                if not queue1.empty():
                    values1 = []
                    while not queue1.empty():
                        values1.append(queue1.get())
                    ax1 = fig.add_subplot(211)
                    Grid_points_u = np.arange(len(u_star))
                    ax1.plot(Grid_points_u, u_star, label='u_star',color='red')
                    ax1.set_xlabel('Grid Points')
                    ax1.set_ylabel('u_star')
                    ax1.legend()

                if not queue2.empty():
                    values = []
                    while not queue2.empty():
                        values.append(queue2.get())

                    ax2 = fig.add_subplot(212)
                    Grid_points_p = np.arange(len(p_star))
                    ax2.plot(Grid_points_p, p_star, label='p_star',color='blue')
                    ax2.set_xlabel('Grid Points')
                    ax2.set_ylabel('p_star')
                    ax2.legend()

                plt.tight_layout()
                plt.show()

                # Check the elapsed time
                elapsed_time = time.time() - start_time
                # Break the loop after a certain amount of time
                if elapsed_time > 60:  # 60 seconds
                    break

        # Start the thread to plot the data
        plot_thread = threading.Thread(target=plot_data)
        plot_thread.start()


```
